# Biblioteca Seaborn
uma das mais famosas bibliotecas de geração de gráficos que eu considero até mais simples que o matplotlib

```python
# imports desse notebook
import seaborn as sea
import random
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
# gerando dados inicias que vem nativos do seaborn para estudo
dados = sea.load_dataset("tips")
dados.head()
```

# Fazendo nosso primeiro gráfico

```python
# Função joinplot
sea.jointplot(data = dados, x =  "total_bill", y = "tip", kind = 'reg')
```

```python
# Função lmplot
sea.lmplot(data = dados, x = 'total_bill', y = 'tip', col = 'smoker')
```

```python
# gerando mais dados
df = pd.DataFrame()
df['idade'] = random.sample(range(20,100),30)
df['peso'] = random.sample(range(55,150),30)
df.head()
```

```python
sea.lmplot(data = df, x='idade',y='peso')
```

```python
# Função kdeplot
sea.kdeplot(df.idade)
sea.kdeplot(df.peso)
```

```python
sea.distplot(df.peso)
```

```python
# Função rugplot
sea.rugplot(df.idade)
```

```python
# Função boxplot
sea.boxplot(df.idade, color = 'm')
```
