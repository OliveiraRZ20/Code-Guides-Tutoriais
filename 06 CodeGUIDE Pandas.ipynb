# Biblioteca PANDAS
essa é a famosa biblioteca que serve como se fosse um excel só que com python e com muito mais opções e facilidades para analise de dados

```python
# Primeiro de tudo é necessário chamar a bilioteca
import pandas as pd

# será importado também o os para configuração dos arquivos do Notebook
import os
```

# Criando nosso primeiro DataFrame
data frames são o que nós chamamos de tabela de excel, é onde a organização de linhas e colunas acontece de verdade, então em primeiro lugar, é necessário atribuir esse data frame a uma variável e ir adicionando dados nele, desde os nomes das colunas, até os diversos dados que serão utilizado na nossa análise

```python
# para começar podemos utilizar um dicionário como forma de alimentar nosso primeiro
# data frame, então segue abaixo alguns dados:
dados = {'Estado': ['Santa Catarina', 'Rio de Janeiro', 'Tocantins', 'Bahia', 'Minas Gerais'], 
         'Ano': [2004, 2005, 2006, 2007, 2008], 
         'Taxa Desemprego': [1.5, 1.7, 1.6, 2.4, 2.7]}


# então agora podemos criar nosso data frame df_ex com a variavel dados como fontee chamálo
df_ex = pd.DataFrame(dados)
df_ex # esse método de chamar apresenta somente 5 linhas do DataFrame
```

# Funções da bilioteca PANDAS

```python
# Função VALUES
# ele vai retornar o dataset em forma de array (lista)
df_ex.values # essa função não necessita de parentesis no final
```

```python
# Função DTYPES
# ele vai retornar o nome de cada coluna com o tipo de dado contido em cada uma
df_ex.dtypes # essa função não necessita de parentesis no final
```

```python
# Função COLUMNS
# ele vai retornar o nome de cada coluna em formato de lista
df_ex.columns # essa função não necessita de parentesis no final
```

```python
# Função SHAPE
# ele vai retornar o tamanho do nosso dataframe no formato (linhaxcoluna)
df_ex.shape # essa função não necessita de parentesis no final
```

```python
# Função DESCRIBE
# ele retorna um resumo estatístico do DataFrame com analises pre calculadas
df_ex.describe()
```

```python
# Função FILTER
# ele vai permitir você aplicar um filtro no data frame
df_ex.filter(items=['Estado']) # exemplo que retorna coluna chamada "Estado"
# mais a frente será mostrada uma forma mais simples de se aplicar filtros
```

```python
# Função SUM
# ele retorna todas as colunas concatenada independente do tipo delas
df_ex.sum()
```

```python
# Função ISNA
# ele retorna o data frame só que substituindo os valores por valores booleanos para
# verificar a presença de valores Nulos (NaN), então se tiver algum True, significa que
# ali tem um valor nulo que precisa ser conscertado
print(df_ex.isna())
print("|=============|")
# para uma analise mais eficiente e rápida, pode se cominar a função ISNA com a SUM
# para se obter uma soma de todos os valores nulos do DataFrame
print(df_ex.isna().sum())
```

# Indexação e filtração de DataFrames
antes de continuarmos com as funções do PANDAS é necessário entender da indexação de dataFrames, então a partir daqui carregaremos mais um dataframe para melhor visualização de algumas funções

DISCLAIMER: lembre-se que a partir do entendimento dessa indexação é possivel aplicar com praticamente todas as funções do PANDAS

```python
# Carregando o dataframe
meu_diretorio = os.path.abspath(".")
path_df = os.path.join(meu_diretorio, f"archives\\dataset.csv") # mude em caso de erro
df_ex1 = pd.read_csv(path_df)
```

```python
# Para se imprimir somente uma coluna do data frame é o mesmo processo das listas python
# só que agora utilizaremos os nomes das colunas
# lembrando que o python interpreta a diferença de "A" e "a", então tome cuidado!!
print(df_ex1['Data_Pedido'])

# para imprimir mais de uma coluna é só colocar a lista de colunas na indexação
print("|================|\n")
print(df_ex1[["Data_Pedido", "Pais"]]) # lembre se de usar dois colchetes para dar certo
```

```python
# Nota: também é possivel chamar colunas utilizando somente um ponto e o nome da coluna
df_ex1.ID_Pedido
```

```python
# Função HEAD e TAIL
# ela serve pra mostrar somente as 5 primeiras ou ultimas linhas de um dataframe
# respectivamente, amplamente utilizada pra visualizar rapidamente
print(df_ex1['Segmento'].head())
print("\n")
print(df_ex1['Segmento'].tail())
```

```python
# Slicing de data frames
# aonde podemos aplicar condições para visualizar linhas especificas
df_ex1[df_ex1['Valor_Venda'] < 2].head() # filtro para taxa desemprego menor que 2
```

```python
# Filtrando as vendas que ocorreram para o segmento de Home Office e na região South
df_ex1[ (df_ex1.Segmento == 'Home Office') & (df_ex1.Regiao == 'South') ].head()
# o sinal & serve como operador AND
```

```python
# Filtrando as vendas que ocorreram para o segmento de Home Office ou região South
df_ex1[(df_ex1.Segmento == 'Home Office') | (df_ex1.Regiao == 'South')].head()
# o sinal | serve como operador OR
```

```python
# Função ISIN
# função interessante pra fazer slicing de dataFrames, pois a mesma permite que voce
# aplice um filtro pra uma lista de valores
df_ex1[df_ex1['Quantidade'].isin([5,7,9,11])].head() # filtro para linhas que tenham os
# valores 5,7,9,11 como valor da coluna Quantidade
```

```python
# Função ENDSWITH e STARTSWITH
# ele aplica o filtro com base na string de uma coluna
df_ex1[df_ex1.Segmento.str.endswith('mer')].head()
# Filtramos o dataframe pela coluna Segmento com valores que terminam com as letras 'mer'
```

```python
# a mesma idaide pode aplicar so que com o STARTWITH
df_ex1[df_ex1.Segmento.str.startswith('Con')].head()
# Filtramos o dataframe pela coluna Segmento com valores que iniciam com as letras 'Con'
```

```python
# Função HEAD e TAIL
# ela serve pra mostrar somente as 5 primeiras ou ultimas linhas de um dataframe
# respectivamente, amplamente utilizada pra visualizar rapidamente
print(df_ex1['Segmento'].head())
print("\n")
print(df_ex1['Segmento'].tail())
```

```python
# Função VALUE_COUNTS
# ele retorna uma lista com todos os valores que tem naquela colunas mas em forma de rank
# para que em primeiro lugar esteja o valor mais repetido naquela coluna
# servindo como forma de se achar a moda de uma coluna
print(df_ex1['Regiao'].value_counts())

# para se obter somente o valor mais repetido é necessário utilizar outra função
# Função INDEX
# é so utilizar e colocar o numero que deseja entre colchetes
print("|===========|")
print(df_ex1['Regiao'].value_counts().index[0])
```

```python
# Função FILLNA
# ele serve para completar valores nulos em uma coluna do dataframe, então com o comando
# isna e sum é possivel localizar a coluna que contém valores nulos
print(df_ex1.isna().sum())

# então com essa informação é possivel resolver o problema da seguinte forma
df_ex1['Quantidade'].fillna(value = 1, inplace=True) # inplace é pra aplicar as alterações
print("\n|============|\n")
print(df_ex1.isna().sum())
```

# Criação de QUERYS em dataframes

```python
# Geramos um novo dataframe apenas com os valores de venda acima da média
df_ex2 = df_ex1.query('Valor_Venda > 766')
df_ex2.head()
```

# Técnica de Group By com DataFrames
Vamos explorar mais a função groupby() pois temos diversas opções de sumarização dos dados de forma simples. No exemplo abaixo uniremos a função groupby() com a função agg() para realiza agregação múltipla.

```python
# Aplicação do group by
df_ex1[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).agg(['mean', 'std', 'count'])
# para melhor entendimento, perceba que a primeira lista de colunas contém todas as envolvidas
# sendo que na função group by, temos somente duas, marcando assim que essas são as colunas
# de agrupamento, deixando assim o valor venda para ser comparado entre as 
# colunas de agrupamento
```

# Tecnica de Split de strings em DataFrames do Pandas
essa é uma técnica que permite utilizar dados do próprio data frame para gerar uma visualização mais limpa, o maior exemplo disso é quando temos uma coluna que contém o ano do pedido em uma tabela de vendas só que concatenada com alguma outra informação

```python
df_ex1['ID_Pedido'].head()
# se voce analisar o retorno do código acima, podemos ver o País, ano e ID do pedido respectivamente
# podemos fazer uma split de trings nessa coluna para tentar extrair o ano do pedido e criar uma coluna
# separada somente para ele
```

```python
df_ex1['ID_Pedido'].str.split("-") # utilizamos o .str pra editar strings no DataFrame
# analisando novamente o retorno no código acima, podemos ver agora a coluna "ID_Pedido" separada pelo
# caracter "-" permitindo assim a captura somente do ano, que é a informação requisitada
```

```python
df_ex1["Ano"] = df_ex1.ID_Pedido.str.split('-').str[1]
df_ex1["Ano"] # assim conseguimos a criação da nossa coluna Ano
```
